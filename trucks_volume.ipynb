```python
import pandas as pd
from pulp import *

# Read in the dataset
df = pd.read_csv('routes_v2.csv')

# Create a list of truck names
trucks = df['truck_id'].unique().tolist()

# Create a dictionary of the volumes of goods for each truck
volumes = {}
for truck in trucks:
    volumes[truck] = df.loc[df['truck_id'] == truck, 'fill_volume'].tolist()

# Create a dictionary of the binary decision variables for each truck and package
x = pulp.LpVariable.dicts('x', ((i, j) for i in trucks for j in range(len(volumes[i]))), cat='Binary')

# Create a pulp optimization problem
problem = pulp.LpProblem('Maximize Volume', LpMaximize)

# Add the objective function to the problem
problem += pulp.lpSum([volumes[i][j] * x[(i, j)] for i in trucks for j in range(len(volumes[i]))]), 'Total Volume'

# Add the constraint that the volume of goods transported by each truck cannot exceed 81
for truck in trucks:
    problem += pulp.lpSum([volumes[truck][j] * x[(truck, j)] for j in range(len(volumes[truck]))]) <= 81, f"truck_id {truck} Constraint"

# Solve the problem
problem.solve()

# Print the status of the problem
print(f"Status: {pulp.LpStatus[problem.status]}")

# Print the maximum volume transported by each truck
for truck in trucks:
    print(f"{truck}: {pulp.value(pulp.lpSum([volumes[truck][j] * x[(truck, j)] for j in range(len(volumes[truck]))]))}")
```
